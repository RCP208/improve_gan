```python
# import libraries
import pickle as pi
import sys
sys.path.append('../')
```

```python
# import tensorboard
from torch.utils.tensorboard import SummaryWriter
```

```python
# import the function class
from scripts.model_clc import MolGen
```

```python
# load data
data = []
with open('../ChEMBL_training/data/database_ChEMBL.csv', "r") as f:
    for line in f.readlines()[1:]:
        data.append(line.split("\n")[0])
```

#### GAN training on the ChEMBL dataset

```python
# set the hyperparameters
lr = 1e-3
bs = 512
steps = 30000
```

```python
# create model
gan_mol = MolGen(data, hidden_dim=128, lr=lr, device="cuda")
```

```python
# create dataloader
loader = gan_mol.create_dataloader(data, batch_size=bs, shuffle=True, num_workers=10)
```

```python
# train model for 30000 steps
gan_mol.train_n_steps(loader, max_step=steps, evaluate_every=1000)
```

```python
# save model
pi.dump(gan_mol, open('checkpoints/gan_mol.pkl', 'wb'))
```

### GAN fine-tuning on the coformer dataset

```python
# load data
data = []
with open('data/database_CCDC.csv', "r") as f:
    for line in f.readlines()[1:]:
        data.append(line.split("\n")[0])
```

```python
# load model
gan_mol = pi.load(open('../ChEMBL_training/checkpoints/gan_1clc_0.2clip_mol_512_0.001_0k.pkl', 'rb'))
```

```python
# set the hyperparameters
lr = 1e-3
bs = 256
steps = 91
```

```python
# set the new learning rate
gan_mol.lr = lr
writer = SummaryWriter(log_dir='runs/training', comment='')
```

```python
# create dataloader
loader = gan_mol.create_dataloader(data, batch_size=bs, shuffle=True, num_workers=10)
```

```python
# train model for 1000 steps
gan_mol.train_n_steps(loader, max_step=steps, evaluate_every=100, logger=writer)
```

```python
# save model
pi.dump(gan_mol, open('checkpoints/gan_1clc_0.2clip_mol_ccdc1_{0}_{1}_{2}k.pkl'.format(bs, '{:.0e}'.format(lr), steps//1000), 'wb'))
```
